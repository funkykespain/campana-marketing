
2.   **Convertimos los datos Trimestrales en Anuales**



```python
# Leer el archivo Excel con los valores trimestrales
productividad_laboral = pd.read_excel("ProductividadLaboral_parcial.xlsx")

# Extraer año de la columna 'Trimestres'
productividad_laboral['Año'] = productividad_laboral['Trimestres'].str[:4]

# Convertir la columna 'Variación anual de la productividad laboral' a numérico
productividad_laboral['Variación anual de la productividad laboral'] = pd.to_numeric(productividad_laboral['Variación anual de la productividad laboral'], errors='coerce')

# Agrupar por año y calcular la media anual
df_grouped_by_year = productividad_laboral.groupby('Año')['Variación anual de la productividad laboral'].mean().reset_index()

# Renombrar columnas
df_grouped_by_year.columns = ['Año', 'Productividad Laboral']

# Guardar el DataFrame con valores anuales en un nuevo archivo Excel
df_grouped_by_year.to_excel('ProductividadLaboral_final.xlsx', index=False)

print("Se han calculado y guardado los valores anuales de la productividad laboral en un nuevo archivo Excel.")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression, Ridge

pipeline_rf = Pipeline([
    ('regressor', RandomForestRegressor())
])

pipeline_ridge = Pipeline([
    ('scaler', StandardScaler()),
    ('regressor', Ridge())
])
```

```python
# Definir cuadrícula de hiperparámetros para regresión de Ridge
param_grid_ridge = {
    'regressor__alpha': [0.1, 1.0, 10.0]
}

# Definir cuadrícula de hiperparámetros para Random Forest Regressor
param_grid_rf = {
    'regressor__n_estimators': [100, 200, 300],
    'regressor__max_depth': [None, 10, 20]
}
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV

# Realizar GridSearchCV para regresión de Ridge
grid_search_ridge = GridSearchCV(pipeline_ridge, param_grid_ridge, cv=5, scoring='neg_mean_squared_error')
grid_search_ridge.fit(X_train, y_train)

# Realizar GridSearchCV para Random Forest Regressor
grid_search_rf = GridSearchCV(pipeline_rf, param_grid_rf, cv=5, scoring='neg_mean_squared_error')
grid_search_rf.fit(X_train, y_train)

# Los mejores parámetros
sorted(grid_search_rf.best_params_)
```
